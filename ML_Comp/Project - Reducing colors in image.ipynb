```python
import matplotlib.pyplot as plt
%matplotlib inline
from skimage.io import imread,imshow
```

```python
img = imread('Beer.jpg')
```

```python
imshow(img)
```

```python
from sklearn.cluster import KMeans
```

```python
img.shape
```

```python
X = img.reshape(-1,3)/255
```

```python
kmeans = KMeans(n_clusters=3)
```

```python
X.shape
```

```python
kmeans.fit(X)
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
new_X = kmeans.cluster_centers_[kmeans.labels_]
```

```python
image = new_X.reshape(526, 793, 3)
```

```python
imshow(image)
```
